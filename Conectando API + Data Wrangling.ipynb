# Información de pilotos temporada 2023

```python
response = requests.get(url_drivers)
print(response.status_code) #Se consulta el status al momento de conectar con la  API
```

```python
def dob_to_age(driver_dob):
    today = datetime.now()
    delta = today - driver_dob
    return int(delta.days / 365.2425)
```

```python
resp_dict = response.json().get("MRData")
resp_drivers = resp_dict.get("DriverTable")
df = pd.DataFrame(resp_drivers.get("Drivers")) #Consideramos solo la información de Drivers y la enviamos a un dataframe
```

A continuación se convierte la columna dateOfBirth de string a datetime ya que para hacer operaciones con las fechas el tipo de dato string no sirve.

```python
df["dateOfBirth"] =  df["dateOfBirth"].apply(pd.to_datetime)
df.sample(20)
```

Con la columna de fecha de nacimiento como tipo datetime ahora se puede usar apply para crear una nueva columna que muestre la edad de cada piloto según su respectiva fecha de nacimiento.

```python
df["age"] = df["dateOfBirth"].apply(dob_to_age)
#El ordenamiento no funciona como quisiera :(
df.sort_values(by=['age'])
df.sample(20)
#df.info()
```

# Información de la programación de carreras 2023

```python
# Al igual que con los anteriores, llevamos la información a un dataset
resp_s = requests.get(url_schedule)
resp_sque = resp_s.json().get("MRData").get("RaceTable").get("Races")
df_sque = pd.DataFrame(resp_sque)
df_sque.info()
```

```python
#Convertimos la columna round a numeros, para luego ordenar por ronda
df_sque["round"] = pd.to_numeric(df_sque["round"])
df_sque.sort_values(by=['round'])
```

#### Las siguientes columnas presentan al menos un valor "NaN"

* ThirdPractice
* Sprint

```python
df_sque["ThirdPractice"].fillna("no data")
df_sque["Sprint"].fillna("no data")
df_sque
```

```python
#Una función para obtener la siguiente carrera según la fecha en que se esté consultando versus el cronograma 
def next_race(df_sque):
    df_sque = df_sque.reset_index() 
    x = datetime.now()

    for index, row in df_sque.iterrows():
        d1 = datetime.strptime(row['date'], "%Y-%m-%d")
        if(d1 > x):
            delta = x - d1
            return(row['date'], row['raceName'], str(int(delta.days*-1))+" dias restantes")    
```

```python

next_race(df_sque)
```
