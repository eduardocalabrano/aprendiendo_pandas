```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('dataset/Rendimiento academico 8va región 2021.csv',sep=';',encoding = "ISO-8859-1", low_memory=False)
```

```python
df.head()
```

```python
mpl.style.use('seaborn-v0_8-pastel')
```

```python
df.index = df['NOM_COM_RBD']
df = df.drop('NOM_COM_RBD', axis='columns')
df
```

```python
x = df.index
gk = df.groupby('NOM_COM_RBD') ##Se agrupa por comuna
y = gk["PROM_GRAL"].mean() # Se realiza el calculo de la media para cada comuna
y
```

```python
fig, ax = plt.subplots(figsize=(8,4))
prom_comuna = gk["PROM_GRAL"].mean()
ax.bar(gk.columns, prom_comuna)
ax.set_title('Precipitaciones acumuladas desde 1965 a 2002, según el mes')
ax.set_ylabel('Precipitación total (mm.)')
ax.set_xlabel('Mes')
```
