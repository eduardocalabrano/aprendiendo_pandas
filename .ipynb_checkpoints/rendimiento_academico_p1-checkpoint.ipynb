```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('dataset/Rendimiento academico 8va región 2021.csv',sep=';',encoding = "ISO-8859-1", low_memory=False)
```

# Lectura de dataset

```python
df.head()
```

```python
mpl.style.use('seaborn-v0_8-pastel')
```

```python
df.index = df['NOM_COM_RBD']
df
```

# Gráficos con Matplotlib

## 1 - Gráfico de Barras - Promedio general de calificaciones por comuna

```python
gk = df.groupby(df.index) ##Se agrupa por comuna
prom_comuna = gk["PROM_GRAL"].mean() # Se realiza el calculo de la media para cada comuna
prom_comuna
```

```python
fig, ax = plt.subplots(figsize=(15,10))
col_list = prom_comuna.index.values.tolist()
bars = ax.barh(col_list, prom_comuna, align='center')
ax.set_title('Promedio de calificaciones por comuna durante el año 2021 (Región del Bio-bio)')
ax.set_xlabel('Rango de calificaciones')
ax.set_ylabel('Comunas')
ax.bar_label(bars)
#Cosas que quedan al debe:
#1.- Poder iniciar el eje X desde 1 en lugar de 0 ya que la calificación minina es 1.0
#2.- Poder ordenar las comunas de  forma descendente según su promedio para visualizar un ranking
plt.show()
```

## 2 - Gráfico de torta - Porporción de alumnos por género

```python
#GEN_ALU = 0: Sin Información - 1: Masculino - 2: Femenino
g_gen = df.groupby('GEN_ALU')
count_gen = g_gen["GEN_ALU"].count()
count_gen
```

```python
col_list2 = count_gen.index.values.tolist()
for i in range(len(col_list2)):
    if col_list2[i] == 0:
        col_list2[i] = 'Sin Información'
    elif col_list2[i] == 1:
        col_list2[i] = 'Masculino'
    else:
        col_list2[i] = 'Femenino'
```

```python
fig, axg = plt.subplots(figsize=(15,5))
axg.pie(count_gen, labels=col_list2, autopct='%1.1f%%',shadow=True, startangle=90)
axg.set_title('Proporción de alumnos por género')
plt.show()
#Cosas que quedan al debe:
#1.- Indicar el total por cada género independiente del porcentaje
```

## 3 - Diagrama de Dispersión - Relación entre calificaciones y asistencia a clases por alumno

```python
fig, axas = plt.subplots(figsize=(15,5))
x = df['ASISTENCIA']
y = df['PROM_GRAL']
axas.scatter(x, y)
axas.set(xlim=(0, 100), ylim=(1, 7))
axas.set_title('Diagrama de dispersión: Relación entre calificaciones y asistencia a clases')
axas.set_xlabel('Porcentaje de asistencia')
axas.set_ylabel('Promedio final del alumno')
plt.show()
#Cosas que quedan al debe:
#1.- Revisando el dataset veo que existen muchos alumnos con promedio apto para aprobar el nivel pero con un
#porcentaje de asistencia por debajo del 20%, aquí descozco si es porque el colegio no marcó bien la asistencia
#de esos alumnos o por causas relacionadas a la pandemia se les permitió aprobar el curso sin considerar la asistencia.
```

## 2.- Gráfico de Barras - Cantidad de alumnos por género y situación académica

```python
#SIT_FIN = P: Promovido - R: Reprobado - Y: Retirado
```

```python
ax_5 = sns.countplot(x ='GEN_ALU', hue = "SIT_FIN", data = df)
ax_5.bar_label(ax_5.containers[0])
ax_5.bar_label(ax_5.containers[1])
ax_5.bar_label(ax_5.containers[2])
ax_5.set(title="Cantidad de alumnos por género y situación académica al finalizar el año")
ax_5.set(xlabel='Género', ylabel='Cantidad de alumnos(as)')
ax_5.set(xticks=([0, 1]))
ax_5.set(xticklabels = (['Masculino', 'Femenino']))
#Cosas que quedan al debe:
#1.- Poder dejar clara la leyenda de SIT_FIN
```
