# Abstract

En Chile, la educación “obligatoria” se divide en el nivel básico y nivel medio, con una duración de 8 y 4 años respectivamente. Iniciándose en 1er nivel básico los alumnos deben tener a la fecha de inicio del calendario escolar una edad mínima de 6 años, por lo que el 1er año de enseñanza media normalmente lo inician a los 14 años, egresando con 17-18 años siempre que no reprueben ningún curso.

En lo que respecta a los centros educacionales existe una categorización que hasta el día de hoy genera debates y dudas sobre el nivel y calidad del contenido educativo que entrega cada uno. Podemos encontrar a los colegios municipales (administración 100% pública), los colegios particulares-subvencionados (administración privada pero con un porcentaje de parte del estado) y las entidades privadas. Además, existen otras formas de categorizar los centros educativos, tales como los “técnico profesional”, escuelas “rurales”, enseñanza para adultos, escuelas para alumnos con necesidades educativas especiales, etc.

Con lo mencionado anteriormente, se puede deducir que existen muchas variables que determinan el rendimiento académico de los alumnos y es importante para cualquier gobierno conocer el escenario que cada año se presenta a nivel país, mismo caso para cada gobierno regional ya que según los datos que entrega cada colegio al finalizar un año escolar, se puede determinar hacia donde es necesario movilizar fondos públicos, gestionar autorizaciones de funcionamiento, entre otros.

En el país, dentro del ministerio de educación existe el “Centro de Estudios Mineduc” (CEM), quien se encarga de entregar los datos año a año sobre rendimiento académico a nivel país.

Durante la duración del curso y los diferentes entregables se trabajará un dataset de rendimiento académico pero únicamente de la 8va región (Bio-Bio) y solo abarca el año 2021. Con estos datos se buscará responder a algunas interrogantes que surgen en la medida que uno revisa el dataset, tales como: “¿La educación particular obtiene mejores resultados que la educación municipal o particular subvencionada?”, “¿Existen personas de tercera edad (mayores de 60 años) que se hayan matriculado en algún colegio durante el 2021?”, “¿Tienen las escuelas rurales mejores resultados que las escuelas en ciudad?” y “¿Existe una importante diferencia en los resultados académicos por género?”. Estas preguntas serán resueltas mediante la utilización de distintas herramientas de visualización de datos donde la correcta interpretación de estas serán la clave para llegar a una acertada conclusión.

```python
df.loc[df['COD_DEPE2'] == 1, 'dependencia'] = 'Municipal'
df.loc[df['COD_DEPE2'] == 2, 'dependencia'] = 'Particular Sub.'
df.loc[df['COD_DEPE2'] == 3, 'dependencia'] = 'Particular Pagado'
df.loc[df['COD_DEPE2'] == 4, 'dependencia'] = 'Particular Sub.'
df.loc[df['COD_DEPE2'] == 5, 'dependencia'] = 'Municipal'

df.loc[df['GEN_ALU'] == 1, 'Género'] = 'Masculino'
df.loc[df['GEN_ALU'] == 2, 'Género'] = 'Femenino'

plt.figure(figsize =(12, 3))
sns.set_style('whitegrid')
ax_s = sns.catplot(data=df, x="dependencia", y="PROM_GRAL", hue="Género", kind="box")
ax_s.set(xlabel='Tipo de dependencia', ylabel='Promedio General')
#ax_s.set(yticks=[1,2,3,4,5,5.3,5.5,6,6.2,6.5,6.8,7])
ax_s.set(yticks=[1,2,3,3.5,4,4.5,5,5.4,6,7])
ax_s.set(title="Boxplot - Diferencia de rendimiento académico por tipo de dependencia y género")
ax_s.set(ylim=(1, 7.5))
plt.xticks(rotation=80)
```

## Interpretación del boxplot

En Chile y gran parte del mundo siempre se debate sobre si la educación pagada asegura mejores resultados académicos. En este diagrama boxplot podemos observar lo siguiente:

1.- Si bien la mediana para los 3 tipos de dependencia está por sobre la nota 6.0, se aprecia un ascenso en la medida que el colegio recibe menos intervención municipal / estatal. Para colegios municipales la mediana es 6,0 (Hombres) y 6,1 (Mujeres), en particulares subencionados es 6,2 (Hombres) y 6,3 (Mujeres) y en el caso de particulares pagados es 6,5 (Hombres) y 6,6 (Mujeres).

2.- Sobre el 50% central de los datos:

    - En el caso de los colegios municipales, el 50% central de los datos se concentra en promedios finales entre 5,3 a 6,5 para hombres y entre 5,5 a 6,6 para mujeres.
    - En el caso de colegios particulares subvencionados, el escenario es entre 5,6 a 6,7 para hombres y 5,8 a 6,7 en el caso de las mujeres.
    - Para los colegios particulares pagados se puede ver que el caso de los hombres está entre 6,0 a 6,8 y en mujeres entre 6,2 a 6,8
    
3.- Al revisar el minimo inlier la diferencia es mucho mayor:

    - En colegios municipales vemos para hombres la califiación final en 3,5 y para mujeres 3,9
    - En particulares subvencionados para hombres un 4,0 y mujeres 4,5
    - Por último en particulares pagados vemos el minimo inlier con 4,9 para hombres y 5,4 en mujeres

**Conclusión:** Se evidencia una diferencia en el rendimiento académico según el tipo de dependencia del colegio y a su vez se logra evidenciar mejores resultados finales en el género femenino por sobre el masculino

## Interpretación de histograma

Es sabido que una no menor parte de la población joven en la década del 60 y 70 no asistía al colegio o lo abandonaron sin dar término formal a sus estudios. En el histograma antes presentado, se evidencia la cantidad de personas de esa época que hoy son considerados "tercera edad" y que decidieron reincorporarse al sistema educativo. De este gráfico podemos concluir que en la región del Bio-Bio se matricularon un total de 23 personas entre los 60 y 80 años, de las cuales 17 son mujeres y sólo 6 son hombres.
